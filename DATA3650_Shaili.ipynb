```python
#A certain number of items need to be packed into boxes, take the total number of items and the number of items 
#that can fit into each boxes as input. how many boxes will be needed and how many items will be left?


total_items = int(input("Enter the total number of items: "))
items_per_box = int(input("Enter how many items fit in each box: "))


num_boxes = total_items // items_per_box

# Calculating the leftover items
leftover = total_items % items_per_box


print(f"You will need {num_boxes} full boxes.")
print(f"There will be {leftover} items left over.")
```
